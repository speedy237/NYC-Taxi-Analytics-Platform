# Architecture Lakehouse - NYC Taxi Data Portfolio

## initialisation de spark sur le catalogue iceberg

```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("NYC_Trip_Taxi")

    # EXTENSION ICEBERG (CRITIQUE)
    .config(
        "spark.sql.extensions",
        "org.apache.iceberg.spark.extensions.IcebergSparkSessionExtensions"
    )

    # CATALOG ICEBERG REST
    .config("spark.sql.catalog.iceberg", "org.apache.iceberg.spark.SparkCatalog")
    .config("spark.sql.catalog.iceberg.type", "rest")
    .config("spark.sql.catalog.iceberg.uri", "http://rest:8181")
    .config("spark.sql.catalog.iceberg.warehouse", "s3://warehouse/")

    # MINIO
    .config("spark.sql.catalog.iceberg.io-impl", "org.apache.iceberg.aws.s3.S3FileIO")
    .config("spark.sql.catalog.iceberg.s3.endpoint", "http://minio:9000")
    .config("spark.sql.catalog.iceberg.s3.path-style-access", "true")
    .config("spark.sql.catalog.iceberg.s3.access-key-id", "minioadmin")
    .config("spark.sql.catalog.iceberg.s3.secret-access-key", "minioadmin123")

    .getOrCreate()
)
```

## Creation des namespace pour l'architecture en medaillons
